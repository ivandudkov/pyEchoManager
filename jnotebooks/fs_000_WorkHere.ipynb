```python
%load_ext autoreload
%autoreload 2
import os
import sys
if not '..' in sys.path:
    sys.path.append('..')

from filemanager.fileset import PDSFileset
from utilities.file_ext_search import file_ext_search
```

```python
fs_folder = r'D:\aa_yandexcloud\aa_cloudmbesproceessing\aa_abp49_processing\assignsvp\filesets'
data_folder = r'D:\zz_ABP49_PDSData\PDS Projects\AkademikBorisPetrov_049\LogData'
data_copy_to = r'C:\PDS Projects\ABP_49\LogData'
```

```python
filesets = file_ext_search(extension='.sub', path=fs_folder)

for idx, fileset in enumerate(filesets):
    print(f'{idx}   {fileset}')
```

```python
fs_list = []

for fileset in filesets[84:85]:
    print(fileset)
    fileset_obj = PDSFileset()
    fileset_obj.readfileset(path=fileset)
    fs_list.append(fileset_obj)
```

```python
fs_size = 0
for fs in fs_list:
    fs.matchfiles(data_folder)
    fs.deleteunmached()
    fs_size += fs.get_size()/2**30
print(f'Размер файлов: {fs_size:.2f}ГБ')
```

```python
import time
proj_dir = r'C:\PDS Projects\ABP_49'

t0 = time.time()
start_time = time.strftime('%A %B %Y %H:%M:%S', time.localtime(t0))

print(f'Copying has started at {start_time}')
count = 0
for fs in fs_list:
    t1 = time.time()
    fs.copyfiles(data_copy_to)
    fs.savefileset(filename=fs.name + '.sub', dirpath=proj_dir)
    t2 = time.time()
    time_to_print = time.strftime('%H:%M:%S', time.localtime(t1 - t2))
    time_to_print2 = time.strftime('%H:%M:%S', time.localtime(t2 - t0))
    print(f'{fs.name} was copied. {count} of {len(fs_list)}')
    print(f'time elapsed: {time_to_print}. Overall time spent: {time_to_print2}')
    count += 1
```
