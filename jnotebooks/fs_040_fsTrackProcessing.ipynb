```python
path = 'Z:\\ABP_50_TTR-21\\zzz_Навигация2йБлок\\Raw\\fs_tr_2ndBlock_005_2022-07-29.csv'
output = 'Z:\\ABP_50_TTR-21\\zzz_Навигация2йБлок\\fs_tr_2ndBlock_005_2022-07-29_proc.csv'
from datetime import datetime, timedelta
x = []
y = []
datetime_list = []

with open(path, 'r') as f:
    file_content = f.read()
    lines = file_content.splitlines()
    for line in lines:
        line_content = line.split(',')
        datetime_list.append(line_content[0])
        x.append(line_content[1])
        y.append(line_content[2])
am = 0
overall = 0
with open(output, 'w') as f2:
    f2.write('x,y,Date,h,m,s\n')
    datetime_utc_prev = None
    for index, time in enumerate(datetime_list):
        datetime_utc = datetime.utcfromtimestamp(float(time))
        if datetime_utc_prev == None:
            datetime_utc_prev = datetime_utc
        else:
            dt = datetime_utc - datetime_utc_prev
            if dt < timedelta(0):
                am += 1
                pass
            else:
                overall += 1
                datetime_utc_prev = datetime_utc
                f2.write(f'{x[index]},{y[index]},{datetime_utc.strftime("%Y%m%d")},{datetime_utc.strftime("%H,%M,%S.%f")}\n')
print(f'Num of falses:{am}, num of norm:{overall}')
        
```
