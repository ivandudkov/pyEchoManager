```python
import os
import numpy as np

%load_ext autoreload
%autoreload 2

from assignsvp import assignsvp as asvp
from filemanager import fileset
from cli import cli

def qgislogToFilesets(output_dir, assignedlog):
    filesets = []
    svp_filesets = []
    filesets_dict = {}

    with open(assignedlog, 'r') as csv_file:
        file_content = csv_file.read().splitlines()

        for num, line in enumerate(file_content[1:]):
            line_content = line.split(',')
            file_name = line_content[1]
            fileset_name = line_content[-1]

            if fileset_name in filesets_dict.keys():
                filesets_dict[fileset_name].append(file_name)
            else:
                filesets_dict[fileset_name] = []
                filesets_dict[fileset_name].append(file_name)
                
    for key in filesets_dict.keys():
        fileset_name = key
        filenames = filesets_dict[key]

        fs = fileset.PDSFileset()
        fs.setname(fileset_name)
        fs.addfiles(filenames)
        filesets.append(fs)
        svp_fs = fs.create_svp_filesets()
        svp_filesets.append(svp_fs)
    
    
    for fs in filesets:
        print(f'{fs.name} - {len(fs.files)}')
        fs.savefileset(dirpath=output_dir)
        
    for svp_fs in svp_filesets:
        print(f'{fs.name} - {len(fs.files)}')
        fs.savefileset(dirpath=output_dir)
```
