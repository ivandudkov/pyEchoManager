```python
file = r'D:\aa_yandexcloud\aa_cloudsbpprocessing\abp53_sbp_processing\abp53_SEB_SBP_to_RNF_processing.csv'

def read_filenames(path):
    
    filenames = []
    
    with open(path, 'r') as f1:
        
        file_content = f1.read().splitlines()
        filenames_idx = None
        
        for num, line in enumerate(file_content):
            line_content = line.split(',')
            if num == 0:
                for num2, content in enumerate(line_content):
                    if 'name' in content:
                        filenames_idx = num2
                        break
            else:
                filenames.append(line_content[filenames_idx])          
    
    return filenames
                
filename_sgy = read_filenames(file)            
                  
```

```python
def check_filename(filename, pathlist):
    checked_path = None
    
    for path in pathlist:
        name = os.path.splitext(os.path.basename(path))[0]

        if name+'_ra' == filename:
            if checked_path is None:
                checked_path = path

            else:
                print(f'Warning. There is a duplicate found: {path}\nPrevious found: {checked_path}')
                
    return checked_path

def check_pathlist(pathlist1, pathlist2):
    checked_paths = []
    
    for filename in pathlist1:
        checked_path = check_filename(filename, pathlist2)
        if checked_path is not None:
            checked_paths.append(checked_path)
    return checked_paths 

segy_paths = check_pathlist(filename_sgy, merged)       
```

```python

for num, file in enumerate(segy_paths):
    copy2(file, r'F:\ABP53\ParasoundP70_data\segy_RNF')
    len1 = len(segy_paths)
    print(f'file {num} of {len1} remaining\n')
```
